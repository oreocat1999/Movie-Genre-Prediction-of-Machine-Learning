```python
# import libraris required
import pandas as pd
import numpy as np
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
```

```python
# load data from files
def load_data():
        
    TFdata = pd.read_csv('train_features.tsv', sep='\t')
    TLdata = pd.read_csv('train_labels.tsv', sep='\t')
    VFdata = pd.read_csv('valid_features.tsv', sep='\t')
    VLdata = pd.read_csv('valid_labels.tsv', sep='\t')
    Testdata = pd.read_csv('test_features.csv')
    TFdata = np.array(TFdata)
    TLdata = np.array(TLdata)
    VFdata = np.array(VFdata)
    VLdata = np.array(VLdata)
    Testdata = np.array(Testdata)
    data = TFdata, TLdata, VFdata, VLdata, Testdata
    return data
 
load_data()
```

```python
# pre-process data 
def feature(data):
    data = np.transpose(data)
    return np.transpose(data[5:])

def label(data):
    data = np.transpose(data)
    return np.transpose(data[-1])

def pre_process_VA():
    TFdata, TLdata, VFdata, VLdata, Testdata = load_data()
       
    X_Train = feature(TFdata)
    Y_Train = label(TLdata)
    X_Valid = feature(VFdata)
    Y_Valid = label(VLdata)   
    X_Test = feature(Testdata)    
    data = X_Train, Y_Train, X_Valid, Y_Valid, X_Test
    
    return data

def pre_process_V():
    X_Train, Y_Train, X_Valid, Y_Valid, X_Test = pre_process_VA()
    
    def video(data):
        data = np.transpose(data)
        return np.transpose(data[:107])
    
    X_Train = video(X_Train)
    X_Valid = video(X_Valid)
    X_Test = video(X_Test)
    #print(X_Train)
    data = X_Train, Y_Train, X_Valid, Y_Valid, X_Test
    
    return data

def pre_process_A():
    X_Train, Y_Train, X_Valid, Y_Valid, X_Test = pre_process_VA()
    
    def audio(data):
        data = np.transpose(data)
        return np.transpose(data[-20:])
    
    X_Train = audio(X_Train)
    X_Valid = audio(X_Valid)
    X_Test = audio(X_Test)
    #print(X_Train)
    data = X_Train, Y_Train, X_Valid, Y_Valid, X_Test
    
    return data

pre_process_VA()
pre_process_V()
pre_process_A()
```

```python
from nltk import word_tokenize
def tag(data):
    data = np.transpose(data)
    data = np.transpose(data[4])
    return tags_clean(data)

def remove_stopwords(tag):
    tag = word_tokenize(tag)
    removed = [word for word in tag if not word in stop_words]
    tag = (" ").join(removed)
    return tag

def tags_clean(tags):    
    new_tags = []
    
    for tag in tags:
        #tag = re.sub("_", "", tag)
        tag = re.sub("[^a-zA-Z]"," ",tag)
        tag = remove_stopwords(tag)
        new_tags.append(tag)        
   
    return new_tags

def pre_process_tag():
    
    TFdata, TLdata, VFdata, VLdata, Testdata = load_data()  
    
    X_Train = tag(TFdata)
    Y_Train = label(TLdata)    
    X_Valid = tag(VFdata)
    Y_Valid = label(VLdata)    
    X_Test = tag(Testdata)    
    data = X_Train, Y_Train, X_Valid, Y_Valid, X_Test
    
    return data

pre_process_tag()  
```

```python
#https://scikit-learn.org/stable/modules/generated/sklearn.feature_extraction.text.TfidfVectorizer.html
def tags2features():
    
    X_Train, Y_Train, X_Valid, Y_Valid, X_Test = pre_process_tag()
    vectorizer =  TfidfVectorizer()
    
    X_Train = vectorizer.fit_transform(X_Train).toarray()
    X_Valid = vectorizer.transform(X_Valid).toarray()
    X_Test = vectorizer.transform(X_Test).toarray()
    
    data = X_Train, Y_Train, X_Valid, Y_Valid, X_Test
    
    return data
    
tags2features()

```

```python
#https://scikit-learn.org/stable/modules/neighbors.html#nearest-neighbors-classification
#Nearest Neighbors
def k_nearest_neighbors(data):
    from sklearn.preprocessing import StandardScaler
    from sklearn.neighbors import KNeighborsClassifier
    from sklearn.metrics import classification_report
    
    scaler = StandardScaler()
    
    X_Train, Y_Train, X_Valid, Y_Valid, X_Test = data
    
    scaler.fit(X_Train)
    
    X_train = scaler.transform(X_Train)
    X_test = scaler.transform(X_Valid)
    
    classifier = KNeighborsClassifier(n_neighbors=7)
    classifier.fit(X_Train, Y_Train)
    y_pred = classifier.predict(X_Valid)
    #print(y_pred)
    from sklearn.metrics import classification_report
    #print(confusion_matrix(y_test, y_pred))
    print(classification_report(Y_Valid, y_pred))
    Result = classifier.predict(X_Test)
    return Result

k_nearest_neighbors(pre_process_VA())
k_nearest_neighbors(pre_process_V())
k_nearest_neighbors(pre_process_A())
k_nearest_neighbors(tags2features())
    
```

```python
#https://scikit-learn.org/stable/modules/neural_networks_supervised.html
#Naive Bayes
def gaussian_naive_bayes(data):
    from sklearn.naive_bayes import GaussianNB
    from sklearn.metrics import classification_report
    
    X_Train, Y_Train, X_Valid, Y_Valid, X_Test = data
    
    gnb = GaussianNB()
    y_pred = gnb.fit(X_Train, Y_Train).predict(X_Valid)
    result = gnb.fit(X_Train, Y_Train).predict(X_Test)
    
    print(classification_report(Y_Valid, y_pred))
    return result
 
gaussian_naive_bayes(pre_process_VA())
gaussian_naive_bayes(pre_process_V())
gaussian_naive_bayes(pre_process_A())
gaussian_naive_bayes(tags2features())   
```

```python
def logistic_regression(data):
    from sklearn.linear_model import LogisticRegression
    from sklearn.multiclass import OneVsRestClassifier
    from sklearn.metrics import classification_report
    
    X_Train, Y_Train, X_Valid, Y_Valid, X_Test = data
    lr = LogisticRegression()
    clf = OneVsRestClassifier(lr)
    clf.fit(X_Train, Y_Train)
    y_pred = clf.predict(X_Valid)
    result = clf.predict(X_Test)
    print(classification_report(Y_Valid, y_pred))
    
    return result

logistic_regression(pre_process_VA())
logistic_regression(pre_process_V())
logistic_regression(pre_process_A())
logistic_regression(tags2features()) 
```

```python
#https://scikit-learn.org/stable/modules/svm.html#svm-classification
#SVM Gram matrix
def support_vector_machine(data):
    
    from sklearn import svm
    from sklearn.metrics import classification_report
    
    X_Train, Y_Train, X_Valid, Y_Valid, X_Test = data
    clf = svm.SVC(kernel='precomputed')
    
    gram_train = np.dot(X_Train, X_Train.T)
    clf.fit(gram_train, Y_Train)
    gram_valid = np.dot(X_Valid, X_Train.T)
    y_pred = clf.predict(gram_valid)
    
    gram_test = np.dot(X_Test, X_Train.T)
    result= clf.predict(gram_test)
    
    print(classification_report(Y_Valid, y_pred))
    return result

support_vector_machine(pre_process_VA())
support_vector_machine(pre_process_V())
support_vector_machine(pre_process_A())
support_vector_machine(tags2features()) 
```
